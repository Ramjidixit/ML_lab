```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler

#load the iris dataset
iris=load_iris()
X=iris.data
y=iris.target


#standardize the features
scaler=StandardScaler()
X_std=scaler.fit_transform(X)

#perfrom pca
pca=PCA()
pca_result=pca.fit_transform(X)

#plot the explained variance ratio
explained_variance_ratio=pca.explained_variance_ratio_
cummulative_variance_ratio=np.cumsum(explained_variance_ratio)

plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance_ratio) +1), explained_variance_ratio, alpha=0.7, align='center', label='Individual explained variance')
plt.step(range(1, len(cummulative_variance_ratio) +1), cummulative_variance_ratio, where='mid', label='Cummulative explained variance')
plt.xlabel('Principal Components')
plt.ylabel('Explained Variance Ratio')
plt.title('Explained Variance Ratio of Prinipal Components')
plt.legend()
plt.show()
```
